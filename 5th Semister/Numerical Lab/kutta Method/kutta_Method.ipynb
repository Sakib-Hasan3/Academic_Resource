```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def f(x, y):
    return x + y
```

```python
def runge_kutta_4(f, x0, y0, h, n):
    """
    Solve ODE using 4th order Runge-Kutta method
    
    Parameters:
    f: function representing dy/dx = f(x, y)
    x0: initial value of x
    y0: initial value of y
    h: step size
    n: number of steps
    
    Returns:
    x_values: array of x values
    y_values: array of y values
    """
    x_values = [x0]
    y_values = [y0]
    
    x = x0
    y = y0
    
    for i in range(n):
        k1 = h * f(x, y)
        k2 = h * f(x + h/2, y + k1/2)
        k3 = h * f(x + h/2, y + k2/2)
        k4 = h * f(x + h, y + k3)
        
        y = y + (1/6) * (k1 + 2*k2 + 2*k3 + k4)
        x = x + h
        
        x_values.append(x)
        y_values.append(y)
    
    return np.array(x_values), np.array(y_values)
```

```python
x0 = 0
y0 = 1
h = 0.1 # Step size
x_end = 1
```

```python
n = int((x_end - x0) / h)

x_vals, y_vals = runge_kutta_4(f, x0, y0, h, n)
```

```python
print("x values:", x_vals)
print("y values:", y_vals)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, 'o-', label='RK4 Approximation')
plt.title('Solution of dy/dx = x + y using RK4')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
