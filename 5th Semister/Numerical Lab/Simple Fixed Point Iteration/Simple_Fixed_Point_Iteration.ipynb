```python
import math
import matplotlib.pyplot as plt
```

```python
def g(k):
    return math.exp(-10 * k) + 0.4286
```

```python
k0 = 0.05
tolerance = 1e-6
max_iter = 100

```

```python
k_values = [k0]
```

```python
for i in range(1, max_iter + 1):
    k1 = g(k0)
    k_values.append(k1)
    error = abs(k1 - k0)
    if error < tolerance:
        break
    k0 = k1
```

```python
k_estimate = k_values[-1]
print(f"Converged after {len(k_values)-1} iterations.")
print(f"Estimated cooling constant k ≈ {k_estimate:.6f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(len(k_values)), k_values, marker='o', linestyle='-')
plt.axhline(y=k_estimate, color='r', linestyle='--', label=f'Final k = {k_estimate:.6f}')
plt.title('Fixed Point Iteration Convergence')
plt.xlabel('Iteration Number')
plt.ylabel('k value')
plt.grid(True)
plt.legend()
plt.show()
```
